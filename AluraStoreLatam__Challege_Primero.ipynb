````python
# Función para graficar top y bottom productos
def mostrar_top_y_bottom_productos(df, nombre_tienda, top_n=5):
    conteo = df['Producto'].value_counts()

    top = conteo.head(top_n)
    bottom = conteo.tail(top_n)
# Gráfico de los más vendidos
    plt.figure(figsize=(6, 5))
    indices_top = np.arange(len(top))
    plt.bar(top.index, top.values, width=0.4, color='green')
    plt.xticks(indices_top, top.index, rotation=45)
    plt.title(f'{nombre_tienda} - Top {top_n} productos más vendidos')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Gráfico de los menos vendidos
    plt.figure(figsize=(6, 5))
    indices_bottom = np.arange(len(bottom))
    plt.bar(indices_bottom, bottom.values, width=0.4, color='red')
    plt.xticks(indices_bottom, bottom.index, rotation=45)
    plt.title(f'{nombre_tienda} - Top {top_n} productos menos vendidos')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


````

````python
# Aplicar a cada tienda
mostrar_top_y_bottom_productos(tienda, "Tienda 1")
mostrar_top_y_bottom_productos(tienda2, "Tienda 2")
mostrar_top_y_bottom_productos(tienda3, "Tienda 3")
mostrar_top_y_bottom_productos(tienda4, "Tienda 4")
````

````python
#Grafico de Barras de Desempeño por Tienda
import numpy as np
import matplotlib.pyplot as plt

# Datos ya calculados (suponiendo que tienes estas listas)
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
x = np.arange(len(tiendas))
ancho = 0.2

# Normalización (si aún no los tienes normalizados)
def normalizar(lista):
    minimo = min(lista)
    maximo = max(lista)
    return [(x - minimo) / (maximo - minimo) for x in lista]

ingresos = [tienda['Precio'].sum(), tienda2['Precio'].sum(), tienda3['Precio'].sum(), tienda4['Precio'].sum()]
calificaciones = [tienda['Calificación'].mean(), tienda2['Calificación'].mean(), tienda3['Calificación'].mean(), tienda4['Calificación'].mean()]
envios = [tienda['Costo de envío'].mean(), tienda2['Costo de envío'].mean(), tienda3['Costo de envío'].mean(), tienda4['Costo de envío'].mean()]
ventas = [len(tienda), len(tienda2), len(tienda3), len(tienda4)]

ingresos_norm = normalizar(ingresos)
calif_norm = normalizar(calificaciones)
envios_norm = [1 - x for x in normalizar(envios)]  # Inverso
ventas_norm = normalizar(ventas)

# Gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x - 1.5*ancho, ingresos_norm, width=ancho, label='Ingresos', color='blue')
plt.bar(x - 0.5*ancho, calif_norm, width=ancho, label='Calificaciones', color='green')
plt.bar(x + 0.5*ancho, envios_norm, width=ancho, label='Costo Envío (invertido)', color='orange')
plt.bar(x + 1.5*ancho, ventas_norm, width=ancho, label='Ventas', color='purple')
plt.xticks(x, tiendas)
plt.ylabel('Rendimiento Normalizado (0 a 1)')
plt.title('Indicadores de Rendimiento por Tienda')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

````python
# Gráfico de pastel (pie) mostrando la proporción de ingresos por tienda
plt.figure(figsize=(6, 6))
plt.pie(ingresos, labels=tiendas, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'purple', 'orange'])
plt.title('Participación de Ingresos por Tienda')
plt.tight_layout()
plt.show()
````

````python
# Gráfico de dispersión que relaciona costo de envío promedio vs ingresos
plt.figure(figsize=(8, 6))
plt.scatter(envios, ingresos, color='red')
for i, name in enumerate(tiendas):
    plt.annotate(name, (envios[i], ingresos[i]), textcoords="offset points", xytext=(0, 5), ha='center')
plt.title('Dispersión: Costo Promedio de Envío vs Ingresos Totales')
plt.xlabel('Costo Promedio de Envío')
plt.ylabel('Ingresos Totales')
plt.grid(True)
plt.tight_layout()
plt.show()
````

# 8. Extra

Se mapean las ventas de cada tienda y se analiza la distribución geográfica de los productos vendidos.

````python
import matplotlib.pyplot as plt

# Crear figura con subgráficos
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Función para graficar la distribución geográfica de ventas
def graficar_distribucion(df, nombre, ax):
    sc = ax.scatter(df['lon'], df['lat'], c=df['Precio'], cmap='viridis', alpha=0.6)
    ax.set_title(f'Distribución Geográfica de Ventas - {nombre}')
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')
    ax.grid(True)
    plt.colorbar(sc, ax=ax, label='Precio de Venta')

# Generar los gráficos por tienda
graficar_distribucion(tienda, 'Tienda 1', axs[0, 0])
graficar_distribucion(tienda2, 'Tienda 2', axs[0, 1])
graficar_distribucion(tienda3, 'Tienda 3', axs[1, 0])
graficar_distribucion(tienda4, 'Tienda 4', axs[1, 1])

# Ajuste final
plt.tight_layout()
plt.show()
````

**Tienda 1:**  Mayor cobertura y dispersión uniforme

      
*   La Tienda 1 muestra una amplia dispersión geográfica en sus ventas, con presencia en múltiples regiones.
*   Tiene zonas de alta densidad de ventas con precios variados, lo cual indica alta penetración y diversidad de público objetivo.
*   Esta amplitud geográfica probablemente contribuye a sus altos ingresos y ventas totales.

🔎 **Conclusión:** Tienda con buen rendimiento general y alcance nacional.

**Tienda 2**: Buena concentración en regiones clave



*   unque menos dispersa que Tienda 1, presenta clusters geográficos bien definidos donde los precios son relativamente altos.
*   Su concentración geográfica parece estar enfocada en zonas económicamente activas, lo que optimiza costos logísticos y explica su buena calificación.


🔎 **Conclusión**: Tienda con rendimiento sólido en zonas específicas, eficiente operativamente.

**Tienda 3:** Actividad moderada, precios medios




*   La distribución de puntos es menos densa y más localizada.
*   Las ventas parecen estar concentradas en una región limitada, con precios intermedios.
*   Esto puede explicar por qué Tienda 3 tiene un rendimiento ligeramente inferior a las tiendas 1 y 2, aunque no críticamente bajo.



🔎 Conclusión: Tienda con presencia limitada, pero aún con comportamiento aceptable.

**Tienda 4**: Concentración baja y ventas dispersas con precios bajos
Tienda 4 muestra una concentración baja de puntos y una dispersión algo errática.




*   Los colores reflejan en su mayoría precios bajos, lo que se alinea con sus menores ingresos.
*  La poca densidad de ventas sugiere débil presencia en regiones clave, y altos costos logísticos por cobertura dispersa.


🔎 **Conclusión**: Rendimiento inferior, influido negativamente por la distribución geográfica no optimizada.

**📌 Conclusión geográfica:**

*  Los patrones geográficos tienen una relación directa con el rendimiento de las tiendas:
*   Tiendas con presencia más densa y localizada en zonas estratégicas (Tienda 1 y 2) muestran mejor desempeño.


La Tienda 4 tiene una cobertura dispersa y baja densidad, tiene mayores costos de envío, menor penetración de mercado y peor desempeño financiero.
